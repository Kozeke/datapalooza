```python
from snowflake.connector.pandas_tools import write_pandas
```

```python
df_travel_predict = df[['EMAIL','ECONOMIC_SITUATION', 'LIKE_TRAVELLING', 'JOB', 'MARITIAL_STATUS']]
```

```python
df_travel_predict['NEXT_WEEK_HOLIDAY'] = True
df_travel_predict['TRAVELLED_THIS_TIME_LAST_YEAR'] = True
df_travel_predict['TRAVEL_NEXT_WEEK'] = True
```

```python
df_travel_predict
```

```python
df_drop = df_travel_predict.dropna()
df_drop = df_drop.drop_duplicates()
df_drop['ID'] = df_drop.index
```

```python
df_drop
```

```python
df= df.reset_index()
```

```python
success, nchunks, nrows, _ = write_pandas(conn=connection
                                          ,df=df_drop
                                          ,table_name='NEXT_WEEK_TRAVEL_PREDICT'
                                          ,schema='PET_CARING_WEBSITE'
                                          ,auto_create_table=True)
```
